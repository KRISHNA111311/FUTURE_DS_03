```python
import pandas as pd

df = pd.read_csv("data/marketing_funnel.csv")
df.head()
```

```python
df.isnull().sum()     # check missing values
df = df.dropna()      # remove missing rows
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.title()
```

```python
print(df.columns)
```

```python
import pandas as pd

total_leads = df.shape[0]

qualified_leads = df['Lead_Type'].notna().sum()

opportunities = df['Seller_Id'].notna().sum()

customers = df['Won_Date'].notna().sum()

funnel_df = pd.DataFrame({
    'Stage': ['Leads', 'Qualified Leads', 'Opportunities', 'Customers'],
    'Count': [total_leads, qualified_leads, opportunities, customers]
})

funnel_df
```

```python
funnel_df['Conversion_Rate'] = funnel_df['Count'] / funnel_df['Count'].iloc[0]
funnel_df
```

```python
funnel_df['Drop_Off_Rate'] = 1 - funnel_df['Count'] / funnel_df['Count'].shift(1)
funnel_df
```

```python
segment_conv = (
    df.groupby('Business_Segment')['Won_Date']
    .apply(lambda x: x.notna().mean())
    .sort_values(ascending=False)
)

segment_conv
```
